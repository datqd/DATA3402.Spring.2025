```python
class counter:
    def __init__(self,max_val):
        self.max_val=max_val
        self.count=0
        
    def increment(self):
        if self.count<self.max_val:
            self.count+=1
            print(f"Success! Incremented by one. The count is {self.count}.")
        else:
            print("Error, can't increment above max.")
```

```python
class Rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError
        
    def is_point_inside(self, x, y):
        raise NotImplementedError
        
    def overlaps(self, other):
        raise NotImplementedError
    
class Rectangle4(Shape):
    def __init__(self, length, width, x, y):
        Shape.__init__(self,x,y)
        self.__length = length
        self.__width = width

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self):
        num_points=16
        points = []

        points_per_side = num_points // 4

        # Side 1 (bottom edge)
        for i in range(points_per_side):
            x = self.get_x() + (i / (points_per_side - 1)) * self.__length
            y = self.get_y()
            points.append((x, y))

        # Side 2 (right edge)
        for i in range(points_per_side):
            x = self.get_x() + self.__length
            y = self.get_y() + (i / (points_per_side - 1)) * self.__width
            points.append((x, y))

        # Side 3 (top edge)
        for i in range(points_per_side):
            x = self.get_x() + (1 - (i / (points_per_side - 1))) * self.__length
            y = self.get_y() + self.__width
            points.append((x, y))

        # Side 4 (left edge)
        for i in range(points_per_side):
            x = self.get_x()
            y = self.get_y() + (1 - (i / (points_per_side - 1))) * self.__width
            points.append((x, y))
        
        return points
    
    def is_point_inside(self, x, y):
        return self.get_x() <= x <= self.get_x() + self.__length and self.get_y() <= y <= self.get_y() + self.__width
    
    def overlaps(self, other):
        if isinstance(other, Rectangle4):
            # Check for rectangle-rectangle overlap
            return not (self.get_x() + self.__length < other.get_x() or
                        self.get_x() > other.get_x() + other.get_length() or
                        self.get_y() + self.__width < other.get_y() or
                        self.get_y() > other.get_y() + other.get_width())

        elif isinstance(other, Circle4):
            # Check for rectangle-circle overlap (simplified)
            closest_x = max(self.get_x(), min(other.get_x(), self.get_x() + self.__length))
            closest_y = max(self.get_y(), min(other.get_y(), self.get_y() + self.__width))
            distance = math.sqrt((closest_x - other.get_x()) ** 2 + (closest_y - other.get_y()) ** 2)
            return distance <= other.get_radius()

        elif isinstance(other, Triangle4):
            # Check for rectangle-triangle overlap
            return (other.is_point_inside(self.get_x(), self.get_y()) or
                    other.is_point_inside(self.get_x() + self.__length, self.get_y()) or
                    other.is_point_inside(self.get_x(), self.get_y() + self.__width) or
                    other.is_point_inside(self.get_x() + self.__length, self.get_y() + self.__width))

        else:
            return False
    
class Circle4(Shape):
    def __init__(self, radius, x, y):
        Shape.__init__(self,x,y)
        self.__radius=radius

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return math.pi * (self.__radius**2)

    def perimeter(self):
        return math.pi * 2 * self.__radius
    
    def perimeter_points(self):
        num_points=16
        points = []
        
        angle_increment = 2 * math.pi / num_points

        for i in range(num_points):
            angle = i * angle_increment
            x = self.get_x() + self.__radius * math.cos(angle)
            y = self.get_y() + self.__radius * math.sin(angle)
            points.append((x, y))
        
        return points
    
    def is_point_inside(self, x, y):
        return (x - self.get_x()) ** 2 + (y - self.get_y()) ** 2 <= self.__radius ** 2
    
    def overlaps(self, other):
        if isinstance(other, Circle4):
            # Check for circle-circle overlap
            distance = math.sqrt((self.get_x() - other.get_x()) ** 2 + (self.get_y() - other.get_y()) ** 2)
            return distance <= (self.get_radius() + other.get_radius())

        elif isinstance(other, Rectangle4):
            # Check for rectangle-circle overlap
            return other.overlaps(self)

        elif isinstance(other, Triangle4):
            # Check for circle-triangle overlap
            for point in other.perimeter_points():
                if self.is_point_inside(point[0], point[1]):
                    return True
            return False

        else:
            return False
    
class Triangle4(Shape):
    def __init__(self,side,x,y):
        Shape.__init__(self,x,y)
        self.__side=side
        
    def get_side(self):
        return self.__side

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return 0.5 * (self.__side**2)

    def perimeter(self):
        tri_diag=(2 * (self.__side**2))**0.5
        return (2 * self.__side) + tri_diag
    
    def perimeter_points(self):
        num_points = 16
        points = []

        points_per_side = num_points // 3  

        def side_points(start_x, start_y, end_x, end_y, num_points):
            side_points = []
            for i in range(num_points):
                x = start_x + (i / (num_points - 1)) * (end_x - start_x)
                y = start_y + (i / (num_points - 1)) * (end_y - start_y)
                side_points.append((x, y))
            return side_points

        # The three vertices of the isosceles right triangle
        vertex1 = (self.get_x(), self.get_y())  # Bottom-left corner
        vertex2 = (self.get_x() + self.__side, self.get_y())  # Bottom-right corner
        vertex3 = (self.get_x() + self.__side / 2, self.get_y() + self.__side)  # Top corner (height)

        side1_points = side_points(vertex1[0], vertex1[1], vertex2[0], vertex2[1], points_per_side)  
        side2_points = side_points(vertex2[0], vertex2[1], vertex3[0], vertex3[1], points_per_side)  
        side3_points = side_points(vertex3[0], vertex3[1], vertex1[0], vertex1[1], points_per_side) 

        points.extend(side1_points)
        points.extend(side2_points)
        points.extend(side3_points)

        return points
    
    def is_point_inside(self, x, y):
        def sign(x1, y1, x2, y2, x3, y3):
            return (x - x3) * (y1 - y3) - (x1 - x3) * (y - y3)

        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = self.get_x() + self.__side, self.get_y()
        x3, y3 = self.get_x() + self.__side / 2, self.get_y() + self.__side

        d1 = sign(x, y, x1, y1, x2, y2)
        d2 = sign(x, y, x2, y2, x3, y3)
        d3 = sign(x, y, x3, y3, x1, y1)

        has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
        has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

        return not (has_neg and has_pos)  # Point is inside if signs are consistent
    
    def overlaps(self, other):
        if isinstance(other, Triangle4):
            # Check for triangle-triangle overlap
            for point in self.perimeter_points():
                if other.is_point_inside(point[0], point[1]):
                    return True
            for point in other.perimeter_points():
                if self.is_point_inside(point[0], point[1]):
                    return True
            return False

        elif isinstance(other, Circle4):
            # Check for circle-triangle overlap
            return other.overlaps(self)

        elif isinstance(other, Rectangle4):
            # Check for rectangle-triangle overlap
            return other.overlaps(self)

        else:
            return False
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        if x1 == x2:
            for y in range(min(y1, y2), max(y1, y2) + 1):
                self.set_pixel(x1, y, **kargs)
            
        else:
            slope = (y2 - y1) / (x2 - x1)
            for y in range(y1, y2 + 1):

                x = int(slope * (y - y1) + x1)  
                self.set_pixel(x, y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
class Shape:
    def __init__(self, x, y,**kwargs):
        self._x = x
        self._y = y
        self.kwargs=kwargs
        
    def __repr__(self):
        raise NotImplementedError

    def get_x(self):
        raise NotImplementedError

    def get_y(self):
        raise NotImplementedError

    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError
        
    def is_point_inside(self, x, y):
        raise NotImplementedError
        
    def overlaps(self, other):
        raise NotImplementedError
        
    def paint(self, canvas):
        pass
    
class Rectangle5(Shape):
    def __init__(self, length, width, x, y,**kwargs):
        Shape.__init__(self,x,y,**kwargs)
        self.__length = length
        self.__width = width
        
    def __repr__(self):
        return f"Rectangle5({repr(self.__length)}, {repr(self.__width)}, {repr(self._x)}, {repr(self._y)}, **{repr(self.kwargs)})"

    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def perimeter_points(self):
        num_points=16
        points = []

        points_per_side = num_points // 4

        # Side 1 (bottom edge)
        for i in range(points_per_side):
            x = self.get_x() + (i / (points_per_side - 1)) * self.__length
            y = self.get_y()
            points.append((x, y))

        # Side 2 (right edge)
        for i in range(points_per_side):
            x = self.get_x() + self.__length
            y = self.get_y() + (i / (points_per_side - 1)) * self.__width
            points.append((x, y))

        # Side 3 (top edge)
        for i in range(points_per_side):
            x = self.get_x() + (1 - (i / (points_per_side - 1))) * self.__length
            y = self.get_y() + self.__width
            points.append((x, y))

        # Side 4 (left edge)
        for i in range(points_per_side):
            x = self.get_x()
            y = self.get_y() + (1 - (i / (points_per_side - 1))) * self.__width
            points.append((x, y))
        
        return points
    
    def is_point_inside(self, x, y):
        return self.get_x() <= x <= self.get_x() + self.__length and self.get_y() <= y <= self.get_y() + self.__width
    
    def overlaps(self, other):
        if isinstance(other, Rectangle4):
            # Check for rectangle-rectangle overlap
            return not (self.get_x() + self.__length < other.get_x() or
                        self.get_x() > other.get_x() + other.get_length() or
                        self.get_y() + self.__width < other.get_y() or
                        self.get_y() > other.get_y() + other.get_width())

        elif isinstance(other, Circle4):
            # Check for rectangle-circle overlap (simplified)
            closest_x = max(self.get_x(), min(other.get_x(), self.get_x() + self.__length))
            closest_y = max(self.get_y(), min(other.get_y(), self.get_y() + self.__width))
            distance = math.sqrt((closest_x - other.get_x()) ** 2 + (closest_y - other.get_y()) ** 2)
            return distance <= other.get_radius()

        elif isinstance(other, Triangle4):
            # Check for rectangle-triangle overlap
            return (other.is_point_inside(self.get_x(), self.get_y()) or
                    other.is_point_inside(self.get_x() + self.__length, self.get_y()) or
                    other.is_point_inside(self.get_x(), self.get_y() + self.__width) or
                    other.is_point_inside(self.get_x() + self.__length, self.get_y() + self.__width))

        else:
            return False
        
    def paint(self, canvas):
        char = self.kwargs.get('char', '*')
        canvas.v_line(self._x, self._y, self.__length, **self.kwargs)
        canvas.v_line(self._x, self._y + self.__width, self.__length, **self.kwargs)
        canvas.h_line(self._x, self._y, self.__width, **self.kwargs)
        canvas.h_line(self._x + self.__length, self._y, self.__width, **self.kwargs)
    
class Circle5(Shape):
    def __init__(self, radius, x, y,**kwargs):
        Shape.__init__(self,x,y,**kwargs)
        self.__radius=radius
        
    def __repr__(self):
        return f"Circle5({repr(self.__radius)}, {repr(self._x)}, {repr(self._y)}, **{repr(self.kwargs)})"

    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return math.pi * (self.__radius**2)

    def perimeter(self):
        return math.pi * 2 * self.__radius
    
    def perimeter_points(self):
        num_points=16
        points = []
        
        angle_increment = 2 * math.pi / num_points

        for i in range(num_points):
            angle = i * angle_increment
            x = self.get_x() + self.__radius * math.cos(angle)
            y = self.get_y() + self.__radius * math.sin(angle)
            points.append((x, y))
        
        return points
    
    def is_point_inside(self, x, y):
        return (x - self.get_x()) ** 2 + (y - self.get_y()) ** 2 <= self.__radius ** 2
    
    def overlaps(self, other):
        if isinstance(other, Circle4):
            # Check for circle-circle overlap
            distance = math.sqrt((self.get_x() - other.get_x()) ** 2 + (self.get_y() - other.get_y()) ** 2)
            return distance <= (self.get_radius() + other.get_radius())

        elif isinstance(other, Rectangle4):
            # Check for rectangle-circle overlap
            return other.overlaps(self)

        elif isinstance(other, Triangle4):
            # Check for circle-triangle overlap
            for point in other.perimeter_points():
                if self.is_point_inside(point[0], point[1]):
                    return True
            return False

        else:
            return False
        
    def paint(self, canvas):
        char = self.kwargs.get('char', '*')
        points = self.perimeter_points()

        for x, y in points:
            canvas.set_pixel(int(x), int(y), **self.kwargs)
    
class Triangle5(Shape):
    def __init__(self,side,x,y,**kwargs):
        Shape.__init__(self,x,y,**kwargs)
        self.__side=side
        
    def __repr__(self):
        return f"Triangle5({repr(self.__side)}, {repr(self._x)}, {repr(self._y)}, **{repr(self.kwargs)})"
        
    def get_side(self):
        return self.__side

    def get_x(self):
        return self._x

    def get_y(self):
        return self._y

    def area(self):
        return 0.5 * (self.__side**2)

    def perimeter(self):
        tri_diag=(2 * (self.__side**2))**0.5
        return (2 * self.__side) + tri_diag
    
    def perimeter_points(self):
        num_points = 16
        points = []

        points_per_side = num_points // 3  

        def side_points(start_x, start_y, end_x, end_y, num_points):
            side_points = []
            for i in range(num_points):
                x = start_x + (i / (num_points - 1)) * (end_x - start_x)
                y = start_y + (i / (num_points - 1)) * (end_y - start_y)
                side_points.append((x, y))
            return side_points

        # The three vertices of the isosceles right triangle
        vertex1 = (self.get_x(), self.get_y())  # Bottom-left corner
        vertex2 = (self.get_x() + self.__side, self.get_y())  # Bottom-right corner
        vertex3 = (self.get_x() + self.__side / 2, self.get_y() + self.__side)  # Top corner (height)

        side1_points = side_points(vertex1[0], vertex1[1], vertex2[0], vertex2[1], points_per_side)  
        side2_points = side_points(vertex2[0], vertex2[1], vertex3[0], vertex3[1], points_per_side)  
        side3_points = side_points(vertex3[0], vertex3[1], vertex1[0], vertex1[1], points_per_side) 

        points.extend(side1_points)
        points.extend(side2_points)
        points.extend(side3_points)

        return points
    
    def is_point_inside(self, x, y):
        def sign(x1, y1, x2, y2, x3, y3):
            return (x - x3) * (y1 - y3) - (x1 - x3) * (y - y3)

        x1, y1 = self.get_x(), self.get_y()
        x2, y2 = self.get_x() + self.__side, self.get_y()
        x3, y3 = self.get_x() + self.__side / 2, self.get_y() + self.__side

        d1 = sign(x, y, x1, y1, x2, y2)
        d2 = sign(x, y, x2, y2, x3, y3)
        d3 = sign(x, y, x3, y3, x1, y1)

        has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
        has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

        return not (has_neg and has_pos)  # Point is inside if signs are consistent
    
    def overlaps(self, other):
        if isinstance(other, Triangle4):
            # Check for triangle-triangle overlap
            for point in self.perimeter_points():
                if other.is_point_inside(point[0], point[1]):
                    return True
            for point in other.perimeter_points():
                if self.is_point_inside(point[0], point[1]):
                    return True
            return False

        elif isinstance(other, Circle4):
            # Check for circle-triangle overlap
            return other.overlaps(self)

        elif isinstance(other, Rectangle4):
            # Check for rectangle-triangle overlap
            return other.overlaps(self)

        else:
            return False
        
    def paint(self, canvas):
        char = self.kwargs.get('char', '*')
        points = self.perimeter_points()

        for i in range(len(points)):
            x1, y1 = points[i]
            x2, y2 = points[(i + 1) % len(points)] 
            canvas.line(int(x1), int(y1), int(x2), int(y2), **self.kwargs)
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        """Add a shape to the drawing."""
        self.shapes.append(shape)

    def remove_shape(self, shape):
        """Remove a shape from the drawing."""
        self.shapes.remove(shape)

    def modify_shape(self, shape, **kwargs):
        """Modify a shape's properties."""
        if isinstance(shape, Rectangle5):
            if 'x' in kwargs:
                shape._x = kwargs['x']
            if 'y' in kwargs:
                shape._y = kwargs['y']
            if 'length' in kwargs:
                shape._Rectangle5__length = kwargs['length']
            if 'width' in kwargs:
                shape._Rectangle5__width = kwargs['width']
        elif isinstance(shape, Circle5):
            if 'x' in kwargs:
                shape._x = kwargs['x']
            if 'y' in kwargs:
                shape._y = kwargs['y']
            if 'radius' in kwargs:
                shape._Circle5__radius = kwargs['radius']
        elif isinstance(shape, Triangle5):
            if 'x' in kwargs:
                shape._x = kwargs['x']
            if 'y' in kwargs:
                shape._y = kwargs['y']
            if 'side' in kwargs:
                shape._Triangle5__side = kwargs['side']

    def paint(self, canvas):
        """Paint all shapes onto the canvas."""
        for shape in self.shapes:
            shape.paint(canvas)
            
    def save(self, filename):
        """Save the drawing to a file."""
        with open(filename, "w") as file:
            file.write(f"[{', '.join(repr(shape) for shape in self.shapes)}]")

    @staticmethod
    def load(filename):
        """Load the drawing from a file."""
        with open(filename, "r") as file:
            shapes_repr = file.read()
            shapes_list = eval(shapes_repr)
            drawing = RasterDrawing()
            for shape in shapes_list:
                drawing.add_shape(shape)
            return drawing

def test_save_load():
    canvas = Canvas(60, 60)

    recta3 = Rectangle5(10, 5, 5, 5, char='#')
    recta4 = Rectangle5(15, 10, 20, 20, char='@')
    circle2 = Circle5(8, 30, 30, char='*')

    drawing = RasterDrawing()
    drawing.add_shape(recta3)
    drawing.add_shape(recta4)
    drawing.add_shape(circle2)

    print("Initial Drawing:")
    drawing.paint(canvas)
    canvas.display()

    filename = "drawing.txt"
    drawing.save(filename)

    canvas.clear_canvas()
    loaded_drawing = RasterDrawing.load(filename)

    print("\nLoaded Drawing:")
    loaded_drawing.paint(canvas)
    canvas.display()

test_save_load()
```
